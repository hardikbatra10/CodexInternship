```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/globalAirQuality_updated.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head(5))
```

```python
coloum_name = 'pm25'
average_value = df[coloum_name].mean()
print(f"\nAverage value of column '{coloum_name}': {average_value}")
```

```python
pollutants = ['pm25', 'pm10', 'no2', 'so2', 'o3', 'co']
avg_values = df[pollutants].mean()

plt.figure(figsize=(7,5))
plt.bar(pollutants, avg_values)
plt.title("Average Pollutant Levels")
plt.xlabel("Pollutant")
plt.ylabel("Average Value")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df['temperature'], df['pm25'])
plt.title("PM2.5 vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("PM2.5")
plt.tight_layout()
plt.show()

numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(9,7))
plt.imshow(corr, cmap='viridis')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

1 Bar Chart  Average Pollutant Levels

What we see:

PM10 and Ozone (O) show the highest average levels.

PM2.5 values are moderately high.

SO and CO levels remain low.

The variations between pollutants show different pollution sources.

Insights:

High PM10 indicates significant coarse particulate pollution, commonly caused by construction dust, road dust, and industrial emissions.

Moderately high PM2.5 suggests fine particles from vehicles, combustion, and smokemore dangerous due to deep lung penetration.

Ozone (O) being high may indicate strong photochemical reactions in the atmosphere (common in sunny, warm conditions).

Lower SO and CO suggest less coal burning or fossil-fuel combustion compared to particulate pollution sources.

2 Scatter Plot  PM2.5 vs Temperature

What we see:

Points are spread widely across all temperature ranges.

There is no clear upward or downward trend.

PM2.5 remains high even at varying temperatures.

Insights:

Temperature does not strongly influence PM2.5 levels in this dataset.

Pollution levels may be driven mainly by traffic, industrial activity, or seasonal dust, rather than weather.

The wide spread indicates high variability in air quality, independent of temperature.

This suggests that emission sources dominate more than meteorological effects.

3 Correlation Heatmap  Relationships Between Variables

What we see:

Most pollutant pairs show weak correlations.

There is a noticeable strong correlation between AQI and PM2.5/PM10, which makes sense since AQI is calculated using major pollutants.

SO, NO, CO, temperature, humidity, and wind speed show low correlation with particulate pollutants.

Insights:

Weak pollutant-to-pollutant correlations reveal that different pollutants originate from different sources:

PM10/PM2.5  Dust, vehicles, industries

NO  Traffic emissions

O  Sunlight-driven chemical reactions

SO  Coal/industrial combustion

Weather conditions (temperature, humidity, wind speed) do not significantly affect pollutant concentration in the dataset.

High correlation of AQI with PM2.5 and PM10 confirms that particulate matter is the primary driver of air quality.

 Overall Conclusions

Particulate matter (PM10 & PM2.5) is the dominant form of pollution in the dataset.

Temperature does not significantly alter PM2.5 concentration, indicating pollution is mainly source-driven rather than climate-driven.

AQI is heavily influenced by PM2.5 and PM10, making these pollutants the main contributors to poor air quality.

The 5 added Indian cities show different pollution characteristics, with cities like Delhi and Kolkata contributing very high PM2.5 and PM10 values.

The heatmap proves that pollutants behave independently, meaning air pollution comes from multiple types of human activity.
